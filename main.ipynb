```python
df_circular_mtl_use = pd.read_csv("./Dataset_CE/Circular_material_use_rate.csv")
```

```python
df_circular_mtl_use_de = df_circular_mtl_use[df_circular_mtl_use["geo"] == "DE"]
```

```python
df_circular_mtl_use_eu = df_circular_mtl_use[df_circular_mtl_use["geo"] == "EU27_2020"]
```

```python
import pandas as pd
import plotly.express as px

# Merge and prepare the dataset
df_compare = pd.merge(
    df_circular_mtl_use_de,
    df_circular_mtl_use_eu,
    on='TIME_PERIOD',
    suffixes=('_GER', '_EU')
)

# Reshape the data for Plotly (long format)
df_long = pd.melt(
    df_compare,
    id_vars=['TIME_PERIOD'],
    value_vars=['OBS_VALUE_GER', 'OBS_VALUE_EU'],
    var_name='Region',
    value_name='Circular Use Rate (%)'
)

# Clean up region names
df_long['Region'] = df_long['Region'].replace({
    'OBS_VALUE_GER': 'Germany',
    'OBS_VALUE_EU': 'European Union'
})

# Create interactive bar chart
fig = px.bar(
    df_long,
    x='TIME_PERIOD',
    y='Circular Use Rate (%)',
    color='Region',
    barmode='group',
    hover_name='Region',
    hover_data={'TIME_PERIOD': True, 'Circular Use Rate (%)': ':.2f'},
    labels={'TIME_PERIOD': 'Year'},
    title='Germany vs. European Union - Circular Material Use Rates'
)

fig.update_layout(
    xaxis_tickangle=-45,
    template='plotly_white',
    legend_title_text='Region'
)

fig.show()
```
